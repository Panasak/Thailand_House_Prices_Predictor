OVERALL GOAL: Get an understanding for which variables are important, view summary statistics, and visualize the data

```python
# -*- coding: utf-8 -*-
from __future__ import unicode_literals
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('../data_collect/housing_data_thailand_2021.csv')
```

```python
df.head(10)
```

```python
# convert prices to float
df['prices'] = df['prices'].str.replace(',','').astype(int)
```

```python
df['prices'].max()
```

```python
# histogram of prices
fig, ax = plt.subplots(figsize=(12,8))
sns.distplot(df['prices'], ax=ax, bins=1000)
ax.set_xlim(3600,249000001)
plt.show()
```

```python
# change the font to Thai
import matplotlib
import matplotlib.font_manager as fm

matplotlib.font_manager.fontManager.addfont('THSarabunNew.ttf')

matplotlib.rc('font', family = 'TH Sarabun New')
matplotlib.rcParams.update({'font.size': 20})
```

```python
# countplot of status
plt.figure(figsize=(12,8))
sns.countplot(x='status',data=df)
```

```python
#heatmap of the corrolation of evething
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True,cmap='viridis')
```

```python
#convert listing_dates to date time
df['listing_date'] = pd.to_datetime(df['listing_date'])
```

```python
#scatterplot between listing dates and prices
plt.figure(figsize=(12,8))
sns.scatterplot(x='prices',y='listing_date',data=df,)
```

```python
#scatterplot between area and prices
plt.figure(figsize=(12,8))
sns.scatterplot(x='prices',y='areas',data=df,)
```

```python
#boxplot between prices and bed
plt.figure(figsize=(12,8))
sns.boxplot(x='beds',y='prices',data=df)
```

```python
#boxplot between prices and bed
plt.figure(figsize=(12,8))
sns.boxplot(x='baths',y='prices',data=df)
```

```python
# fill in missing subdistrict with ''
def fix_address(address):
    if address.count(",") >= 2:
        return address
    else:
        result = ' , '+address
        return result
```

```python
# apply the def
df['address'] = df['address'].apply(lambda x: fix_address(x))
```

```python
df
```

```python
#seperate the address
df['subdistrict'] = df['address'].apply(lambda x: x.split(',')[0])
df['district'] = df['address'].apply(lambda x: x.split(',')[1])
df['province'] = df['address'].apply(lambda x: x.split(',')[2])
```

```python
pd.set_option('display.max_rows', None)
more_commas = df[df['address'].apply(lambda x: x.count(",")) > 2]['address']
less_commas = df[df['address'].apply(lambda x: x.count(",")) < 2]['address']
```

```python
df['subdistrict'] = more_commas.apply(lambda x: x.split(',')[:1])
df['district'] = more_commas.apply(lambda x: x.split(',')[2])
df['province'] = more_commas.apply(lambda x: x.split(',')[3])
```
