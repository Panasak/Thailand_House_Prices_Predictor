```python
# -*- coding: utf-8 -*-
from __future__ import unicode_literals
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('../data_collect/housing_data_thailand_2021.csv')
```

```python
# fill in missing subdistrict with ''
def fix_address(address):
    if address.count(",") >= 2:
        return address
    else:
        result = ' , '+address
        return result
```

```python
# apply the def
df['address'] = df['address'].apply(lambda x: fix_address(x))
```

```python
df
```

```python
#PROBLEM: SOME PROVES ARE STILL NOT CORRECT
address_split = df['address'].apply(lambda x:x.split(','))
```

```python
# seperate address into subdistricts, districts, and province
sub = []
dis = []
prov = []
for a_l in address_split:
    if a_l.count(',') > 2:
        sub.append(a_l[:2])
        dis.append(a_l[2])
        prov.append(a_l[3])
    else:
        sub.append(a_l[0])
        dis.append(a_l[1])
        prov.append(a_l[2])
df['sub'] = sub
df['dis'] = dis
df['prov'] = prov
```

```python
df[['prices','prov']].groupby('prov').mean().plot(kind='bar',figsize=(16,8))
```
